```python
from text_embedder import ProteinQueryEncoder
import pandas as pd
import torch
from torch.utils.data import Dataset, DataLoader
```

```python
class ProteinQueryDataset(Dataset):
    def __init__(self, csv_file):
        data = pd.read_csv(csv_file)
        self.cell_ids = list(data["cell_id"])
        self.queries = list(data["text"])        
        self.len = len(self.queries)
        
    def __len__(self):
        return self.len

    def __getitem__(self, idx):
        return self.queries[idx], self.cell_ids[idx]
```

```python
model_name = "dmis-lab/biobert-v1.1"
max_length = 128
encoder = ProteinQueryEncoder(model_name)
```

```python
dataset = ProteinQueryDataset("cell_texts.csv")
loader = DataLoader(dataset, batch_size=16)
```

```python
batch = next(iter(loader))
out = encoder(batch)
out.shape
```

```python
_, queries = batch
tok_dict = encoder.tokenizer(
    queries, 
    padding=True, 
    return_tensors='pt', 
)
```
